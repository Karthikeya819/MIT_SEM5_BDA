```python
from pyspark import SparkContext
```

```python
sc = SparkContext('local', 'EndSem2')
```

```python
txt = sc.textFile('file:///home/karthikeya/Desktop/sem5/MIT_SEM5_BDA/Week2/Q5.txt')
```

```python
words = txt.flatMap(lambda x: list(map(str.lower, x.split())))
```

```python
words = words.map(lambda x: (x, 1))
```

```python
words_count = words.reduceByKey(lambda x, y: x + y)
```

```python
words_count.collect()
```

```python
sc.stop()
```
