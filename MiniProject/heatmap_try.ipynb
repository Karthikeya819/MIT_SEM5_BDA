```python
import plotly.express as px
import pandas as pd
```

```python
# Sample data
data = pd.DataFrame({
    'City': ['Delhi', 'Mumbai', 'Chennai'],
    'Latitude': [28.61, 19.07, 13.08],
    'Longitude': [77.23, 72.87, 80.27],
    'AQI': [210, 130, 90],
    'AQI_Bucket': ['Very Poor', 'Moderate', 'Satisfactory']
})

# Use color scale to represent AQI
fig = px.scatter_mapbox(
    data,
    lat="Latitude",
    lon="Longitude",
    hover_name="City",
    hover_data=["AQI", "AQI_Bucket"],
    color="AQI",
    size_max=15,
    zoom=4,
    color_continuous_scale=px.colors.sequential.YlOrRd,
    mapbox_style="carto-positron"
)

fig.write_html("aqi_map.html")
import webbrowser
webbrowser.open("aqi_map.html")
```
